# Chapter 0. 선형성 정의 및 1차 연립방정식의 의미

Linearity : *f*(x) -> function, operation

Lineartiy의 두가지 조건

* 1) superposition : $ f(x_1, x_2) = f(x_1) + f(x_2) $  
* 2) homogeneity : $f(ax) = af(x) $

by 1) and 2)  $ f(a_1x_1 + a_2x_2) = a_1f(x_1) + a_1f(x_2) $


> $ y = mx + n (m\neq0) $ 인 직선이 존재할 때,

> $ m(a_1x_1 + a_2x_2) + n = a_1(mx_1 + n) + a_2(mx_2 + n) $ 이다

따라서 원점을 지나지 않는 직선은 선형성이 있다고 할 수 없다.


differentiation(미분)
> $ \frac{d}{dt}(a_1x_1(t) + a_2x_2(t)) = a_1\frac{d}{dt}x_1(t) + a_2\frac{d}{dt}x_2(t) $

integration(적분)
> $ \int(a_1x(t) + a_2x_2(t))dt = a_1\int x_1(t)dt + a_2\int x_2(t)dt $

Matrix operation(multiplication)


>$ A(X_1 + X_2) = AX_1 + AX_2 $ 이므로

$ A(a_1X_1 + a_2X_2) = a_1AX_1 + a_2AX_2 $

### Basic Notation of Matrix

Vector : V = $(a, b, c)$ -> row vector 
         
 　　　 V = $\begin{bmatrix} a \\ b \\ c \end{bmatrix}$ -> column vector  

Transpose(전치)

　　　　　　　　　　　row  　　<->　　 column

$$\begin{bmatrix} a & b & c \\ d & e & f \end{bmatrix}^T = \begin{bmatrix} a & d \\ b & e \\ c & f \end{bmatrix}$$

Linear Combination(선형 결합)

$$\mathbb{V} = \begin{bmatrix} a_1 \\ b_1 \\ c_1 \end{bmatrix} 　　　
\mathbb{W} = \begin{bmatrix} a_2 \\ b_2 \\ c_2 \end{bmatrix}$$
\
\
$\alpha\mathbb{V} + \beta\mathbb{W} = \alpha\begin{bmatrix} a_1 \\ b_1 \\ c_1 \end{bmatrix} + \beta\begin{bmatrix} a_2 \\ b_2 \\ c_2 \end{bmatrix} = \begin{bmatrix} \alpha a_1 + \beta a_2 \\ \alpha b_1 + \beta b_2 \\ \alpha c_1 + \beta c_2 \end{bmatrix}$, 　　　 $ \begin{bmatrix} a_1 & a_2 \\ b_1 & b_2 \\ c_1 & c_2 \end{bmatrix} \begin{bmatrix} \alpha \\ \beta \end{bmatrix} = \begin{bmatrix} \alpha a_1 + \beta a_2 \\ \alpha b_1 + \beta b_2 \\ \alpha c_1 + \beta c_2 \end{bmatrix} = \alpha\mathbb{V} + \beta\mathbb{W} $

#### Vector 

$\mathbb{V_1} = (a_1, b_1, c_1)^T = \begin{bmatrix} a_1 \\ b_1 \\ c_1 \end{bmatrix} $, 　　　
$\mathbb{V_2} = (a_2, b_2, c_2)^T = \begin{bmatrix} a_2 \\ b_2 \\ c_2 \end{bmatrix} $

\

$\mathbb{V_1} + \mathbb{V_2} = (a_1 + a_2, b_1 + b_2, c_1 + c_2)$

$\mathbb{V_1} - \mathbb{V_2} = (a_1 - a_2, b_1 - b_2, c_1 - c_2)$

\

inner product(내적)

$\mathbb{V_1} \mathbb{V_2} = |\mathbb{V_1}||\mathbb{V_2}| \cos \theta = (a_1a_2 + b_1b_2 + c_1c_2)$

* central ideas(terms)

  - column space : all combinations of the columns
  - row space : all combinations of the rows
  - rank : # of independant columns (or rows)
  - elimination : to find the rank of a matrix
