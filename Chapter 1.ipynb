```python
def Back_Subs(A,f):
    n = f.size
    x = np.zeros(n)            
    x[n-1] = f[n-1]/A[n-1,n-1]    
    for i in range(n-2,-1,-1):    
        sum_ = 0
        for j in range(i+1,n):      
            sum_ = sum_ + A[i,j]*x[j]
        x[i] = (f[i] - sum_)/A[i,i]
    return x
```
